### Distribution plots
* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

```python
tips = sns.load_dataset('tips')  #loading a built in data set
```

```python
sns.distplot(tips['total_bill'])
```

You can think of the Kernel Density Estimation as a smoothed histogram which we can ignore for now.
To know more about KDE visit  [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth).

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

```python
#barplot
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#count plot
sns.countplot(x='sex',data=tips)
```

## factorplot

factorplot is the most general form of a categorical plot. It can take in a **kind** parameter to adjust the plot type:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```
